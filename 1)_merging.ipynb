```python
# Import necessary libraries
import sqlite3
import pandas as pd

# Connect to a SQLite database (or create one)
conn = sqlite3.connect("/content/data.db")
cursor = conn.cursor()
```

```python
accounts = pd.read_sql_query("SELECT * FROM accounts", conn)
campaigns = pd.read_sql_query("SELECT * FROM campaigns", conn)
clients = pd.read_sql_query("SELECT * FROM clients", conn)
outcomes = pd.read_sql_query("SELECT * FROM outcomes", conn)
```

```python
clients.rename(columns={"id": "client_id"}, inplace=True)
clients.head(1)
```

```python
accounts.rename(columns={"id": "account_id"}, inplace=True)
accounts.head(1)
```

```python
client_acc = pd.merge(clients, accounts, on="client_id")
client_acc.head(1)
```

```python
campaigns.rename(columns={"id": "campaign_id"}, inplace=True)
campaigns.head(1)
```

```python
client_acc_camp = pd.merge(client_acc, campaigns, on="account_id")
client_acc_camp.head(1)
```

```python
outcomes.drop(columns=["id"], inplace=True)
outcomes.head(1)
```

```python
merged = pd.merge(client_acc_camp, outcomes, on="campaign_id")
print(merged.shape)
merged.head(1)
```

```python
random_rows = merged.sample(n=3)
random_rows
```

```python
print(merged.isnull().sum(axis = 0))
print(merged.describe())
```

```python
merged.to_csv('merged_data.csv', index=False)
```
