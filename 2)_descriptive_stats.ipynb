```python
import pandas as pd
url = "https://raw.githubusercontent.com/Altusya/Case-Study/main/merged_data.csv"
data = pd.read_csv(url)
data.head(1)
```

```python
duplicateRows = data[data.duplicated(subset="client_id")]
duplicateRows
```

```python
grouped_y = data.groupby('y').size().reset_index(name='count')
grouped_y
```

```python
pivot_job = pd.pivot_table(data=data, columns="job", values="account_id", aggfunc='count')
print("\033[1m  Type of job \033[0m")
print(pivot_job)
print()

pivot_marital = pd.pivot_table(data=data, columns="marital", values="account_id", aggfunc='count')
print("\033[1m  Marital status of the client  \033[0m")
print(pivot_marital)
print()

pivot_education = pd.pivot_table(data=data, columns="education", values="account_id", aggfunc='count')
print("\033[1m  Educational level \033[0m")
print(pivot_education)
print()

pivot_default = pd.pivot_table(data=data, columns="in_default", values="account_id", aggfunc='count')
print("\033[1m  If the person has credit in default \033[0m")
print(pivot_default)
print()

pivot_housing = pd.pivot_table(data=data, columns="housing", values="account_id", aggfunc='count')
print("\033[1m  If the person has taken a Housing Loan  \033[0m")
print(pivot_housing)
print()

pivot_loan = pd.pivot_table(data=data, columns="loan", values="account_id", aggfunc='count')
print("\033[1m  If the person has taken a Personal Loan \033[0m")
print(pivot_loan)
print()

pivot_contact = pd.pivot_table(data=data, columns="contact", values="account_id", aggfunc='count')
print("\033[1m  Preferred contact communication type  \033[0m")
print(pivot_contact)
print()

pivot_campaign = pd.pivot_table(data=data, columns="campaign", values="account_id", aggfunc='count')
print("\033[1m  Number of contacts performed during this campaign to the client \033[0m")
print(pivot_contact)
print()

pivot_poutcome = pd.pivot_table(data=data, columns="poutcome", values="account_id", aggfunc='count')
# ALTERNATIVE: grouped_poutcome = data.groupby('poutcome').size().reset_index(name='count')
print("\033[1m  Outcome of the previous marketing campaign  \033[0m")
print(pivot_poutcome)
print()
```

```python
cross_contact_poutcome = pd.crosstab(data['contact'], data['poutcome'])
cross_contact_poutcome
```

```python
cross_poutcome_y = pd.crosstab(data['poutcome'], data['y'], normalize='index').style.format("{:.2%}")
cross_poutcome_y
```

```python
# for clients over x EUR in balanace
data_vip = data[data['balance'] > 30000]
cross_poutcome_y = pd.crosstab(data_vip['poutcome'], data_vip['y'], normalize='index').style.format("{:.2%}")
cross_poutcome_y
```
